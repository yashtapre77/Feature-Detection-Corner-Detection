```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
img = cv2.imread('building.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

```python
# Shi-Tomasi corner detection
corners = cv2.goodFeaturesToTrack(
    image=gray,
    maxCorners=200,        # Maximum number of corners to return
    qualityLevel=0.15,     # Minimum accepted quality of corners
    minDistance=5         # Minimum distance between detected corners
)

# Convert corners to int
corners = np.round(corners).astype(int)
```

```python
# Draw corners on the image
for corner in corners:
    x, y = corner.ravel()
    cv2.circle(img, (x, y), 4, (0, 255, 0), 20)  # Green filled circle

# Display the image with corners
plt.figure(figsize=(10, 6))
plt.imshow(img)
plt.title("Shi-Tomasi Corner Detection")
plt.axis('off')
plt.show()
```
